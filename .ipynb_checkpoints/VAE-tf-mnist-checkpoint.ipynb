```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import argparse
import os
import cv2
from mpl_toolkits.mplot3d import Axes3D
from tensorflow.examples.tutorials.mnist import input_data
%matplotlib inline
import keras.backend as K
from IPython.display import clear_output
```

```python
# MNIST dataset
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
image_size = 32
num_train = mnist.train.images.shape[0]
batch_size = 128
latent_dim = 2
epochs = 100
learning_rate = 1e-4
beta = 1
C_multiplier = 0.0015
C_limit = 25.0
```

```python
# reset graph
tf.reset_default_graph()

he_init = tf.contrib.layers.variance_scaling_initializer()
kernel_init = he_init
act_fn = tf.nn.relu

# network architecture
with tf.name_scope('encoder') as scope:
    num_filter = 16
    x0 = tf.placeholder(tf.float32, [None, image_size, image_size, 1])
    x = tf.layers.conv2d(inputs = x0, 
                         filters = num_filter, 
                         kernel_size = (3, 3),
                         strides = (2, 2), 
                         padding = 'same',
                         kernel_initializer = kernel_init, 
                         activation = act_fn)
    for i in range(3):
        num_filter = num_filter * 2
        x = tf.layers.conv2d(inputs = x, 
                             filters = num_filter, 
                             kernel_size = (3, 3),
                             strides = (2, 2), 
                             padding = 'same',
                             kernel_initializer = kernel_init, 
                             activation = act_fn)
    shape = K.int_shape(x)
    x = tf.contrib.layers.flatten(x)
    x = tf.layers.dense(x, 256, activation = act_fn)
    
with tf.name_scope('latent') as scope:
    z_mean = tf.layers.dense(x, latent_dim, name = 'z_mean')
    z_log_var = tf.layers.dense(x, latent_dim, name = 'z_log_var')
    dist = tf.distributions.Normal(0., 1.)
    epsilon = dist.sample()
    z_sample = z_mean + tf.exp(0.5 * z_log_var) * epsilon

with tf.name_scope('decoder') as scope:
    x = tf.layers.dense(z_sample, 256, activation = act_fn)
    x = tf.layers.dense(x, shape[1]*shape[2]*shape[3])
    x = tf.reshape(x, [-1, shape[1], shape[2], shape[3]])
    for i in range(3):
        x = tf.layers.conv2d_transpose(inputs = x,
                                       filters = num_filter, 
                                       kernel_size = (3, 3),
                                       strides = (2, 2), 
                                       padding = 'same',
                                       kernel_initializer = kernel_init, 
                                       activation = act_fn)
        num_filter = num_filter // 2
        
    x = tf.layers.conv2d_transpose(inputs = x,
                                   filters = 1, 
                                   kernel_size = (3, 3),
                                   strides = (2, 2), 
                                   padding = 'same',
                                   kernel_initializer = kernel_init)
    x_ = tf.nn.sigmoid(x, name = 'output')

# loss function
step = tf.placeholder(tf.float32)
C = tf.minimum(C_multiplier * step, C_limit)
reconstruction_loss = tf.losses.mean_squared_error(K.flatten(x0), K.flatten(x_))
reconstruction_loss *= image_size * image_size
kl_loss = -0.5 * tf.reduce_sum(
    1 + z_log_var 
    - tf.square(z_mean) 
    - tf.exp(z_log_var))
# loss = reconstruction_loss + beta * tf.abs(kl_loss - C)
loss = reconstruction_loss + beta * kl_loss
optimizer = tf.train.AdamOptimizer(learning_rate)
train_op = optimizer.minimize(loss)
init_op = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
step_value = 0.0
loss_hist = []
loss_recon_hist = []
loss_kl_hist = []
with tf.Session() as sess:
    sess.run(init_op)
    for epoch in range(epochs):
        for i in range(num_train//batch_size):
            train_imgs_sampled = mnist.train.images[i*batch_size:(i+1)*batch_size]
            train_imgs_sampled = np.reshape(train_imgs_sampled, [-1, 28, 28, 1])
            buffer = np.zeros((batch_size, image_size, image_size, 1))
            for i in range(batch_size):
                buffer[i, 2:-2, 2:-2, :] = train_imgs_sampled[i, :, :, :]
            _, loss_value, loss_recon, loss_kl, x_recon = sess.run([
                train_op, loss, reconstruction_loss,  kl_loss, x_], feed_dict = {x0: buffer})
            loss_hist.append(loss_value)
            loss_recon_hist.append(loss_recon)
            loss_kl_hist.append(loss_kl)
        clear_output()
        if (len(loss_hist) < 300):
            plt.figure()
            plt.plot(loss_hist, label = 'total loss')
            plt.plot(loss_recon_hist, label = 'recon loss')
            plt.plot(loss_kl_hist, label = 'kl loss')
            plt.legend()
            plt.grid()
            plt.show()
        else:
            plt.figure()
            plt.plot(loss_hist[-300:], label = 'total loss')
            plt.plot(loss_recon_hist[-300:], label = 'recon loss')
            plt.plot(loss_kl_hist[-300:], label = 'kl loss')
            plt.legend()
            plt.grid()
            plt.show()
        
        plt.figure()
        idx = np.random.randint(0, num_train//batch_size)
        train_imgs_sampled = mnist.train.images[idx*batch_size:(idx+1)*batch_size]
        train_imgs_sampled = np.reshape(train_imgs_sampled, [-1, 28, 28, 1])
        buffer = np.zeros((batch_size, image_size, image_size, 1))
        for i in range(batch_size):
            buffer[i, 2:-2, 2:-2, :] = train_imgs_sampled[i, :, :, :]
        x_recon = sess.run(x_, feed_dict = {x0: buffer})
        to_show = np.zeros((image_size*2, image_size*10))
        for i in range(10):
            to_show[:image_size, i*image_size:(i+1)*image_size] = np.squeeze(buffer[i])
            to_show[image_size:, i*image_size:(i+1)*image_size] = np.squeeze(x_recon[i])
        plt.imshow(to_show, cmap = 'gray')
        plt.show()
        print('Epoch: {}, loss: {}'.format(epoch, np.mean(loss_value)))
        print('==================================')
    save_path = saver.save(sess, "./dsprite_vae.ckpt")
```
