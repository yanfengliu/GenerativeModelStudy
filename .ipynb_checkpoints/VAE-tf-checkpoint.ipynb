```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import argparse
import os
from mpl_toolkits.mplot3d import Axes3D
from tensorflow.examples.tutorials.mnist import input_data
%matplotlib inline
```

```python
# reset graph
tf.reset_default_graph()

# MNIST dataset
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)

num_train = mnist.train.images.shape[0]
original_dim = mnist.train.images.shape[1]

# network parameters
input_shape = (original_dim, )
intermediate_dim = 512
batch_size = 128
latent_dim = 2
epochs = 50

# network architecture
x0 = tf.placeholder(tf.float32, [None, 784])
x = tf.layers.dense(x0, intermediate_dim, activation = tf.nn.relu)
z_mean = tf.layers.dense(x, latent_dim, name = 'z_mean')
z_log_var = tf.layers.dense(x, latent_dim, name = 'z_log_var')
dist = tf.distributions.Normal(0., 1.)
epsilon = dist.sample()
z_sample = z_mean + tf.exp(0.5 * z_log_var) * epsilon
x = tf.layers.dense(z_sample, intermediate_dim, activation = tf.nn.relu)
x_ = tf.layers.dense(x, original_dim, activation = tf.nn.sigmoid)

# loss function
beta = 1.0
reconstruction_loss = tf.losses.mean_squared_error(x0, x_) * original_dim
kl_loss = 0.5 * tf.reduce_sum(
    tf.square(z_mean) + tf.exp(z_log_var) - 1 - z_log_var, 
    axis = -1)
loss = reconstruction_loss + beta * kl_loss
optimizer = tf.train.AdamOptimizer(0.0001)
train_op = optimizer.minimize(loss)
init_op = tf.global_variables_initializer()

sess = tf.InteractiveSession()
sess.run(init_op)
for epoch in range(epochs):
    for i in range(num_train//batch_size):
        batch_xs, batch_ys = mnist.train.next_batch(batch_size)
        _, loss_value = sess.run([train_op, loss], feed_dict = {x0: batch_xs})
    print('Epoch: {}, loss: {}'.format(epoch, np.mean(loss_value)))
```

```python

z_mean_values = sess.run(z_mean, feed_dict = {x0: mnist.test.images})

# 2D plot
y_test = mnist.test.labels.argmax(axis = -1)
y_test = np.reshape(y_test, [10000])
plt.figure(figsize=(12, 10))
plt.scatter(z_mean_values[:, 0], z_mean_values[:, 1], c=y_test)
plt.colorbar()
plt.xlabel("z[0]")
plt.ylabel("z[1]")
plt.show()

# display a 30x30 2D manifold of digits
n = 30
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))
# linearly spaced coordinates corresponding to the 2D plot
# of digit classes in the latent space
grid_x = np.linspace(-4, 4, n)
grid_y = np.linspace(-4, 4, n)[::-1]

for i, yi in enumerate(grid_y):
    for j, xi in enumerate(grid_x):
        z = np.array([[xi, yi]])
        x_decoded = sess.run(x_, feed_dict = {z_sample: z})
        digit = x_decoded[0].reshape(digit_size, digit_size)
        figure[i * digit_size: (i + 1) * digit_size,
               j * digit_size: (j + 1) * digit_size] = digit

plt.figure(figsize=(10, 10))
start_range = digit_size // 2
end_range = n * digit_size + start_range + 1
pixel_range = np.arange(start_range, end_range, digit_size)
sample_range_x = np.round(grid_x, 1)
sample_range_y = np.round(grid_y, 1)
plt.xticks(pixel_range, sample_range_x)
plt.yticks(pixel_range, sample_range_y)
plt.xlabel("z[0]")
plt.ylabel("z[1]")
plt.imshow(figure, cmap='Greys_r')
plt.show()
sess.close()
```
